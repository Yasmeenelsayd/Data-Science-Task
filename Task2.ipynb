```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df= pd.read_csv('train.csv',index_col='ID')
```

```python
df.sort_index(inplace=True)
```

```python
df.head(30)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df['ManufacturingYear'] = df['ManufacturingYear'].str.extract(r'(\b\d{4}\b)', expand=False)
```

```python
df.head(30)
```

```python
df['Duty'] = pd.to_numeric(df['Duty'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.head(30)
```

```python
df['#airbags'] = df['#airbags'].str.extract(r'(\d+\.?\d*)', expand=False)
```

```python
df['#airbags'].unique()
```

```python
df['CylinderCount'] = df['CylinderCount'].str.extract(r'(\d+\.?\d*)', expand=False)
```

```python
df['CylinderCount'].unique()
```

```python
df['CylinderCount'] = pd.to_numeric(df['CylinderCount'], errors='coerce')
```

```python
df['Brand'].unique()
```

```python
brand_mappings = {
    r'.*kia.*': 'Kia',
    r'.*hyundai.*': 'Hyundai',
    r'.*toyota.*': 'Toyota',
    r'.*honda.*': 'Honda',
    r'.*Lexus.*': 'Lexus',
    r'.*Opel.*': 'Opel',
    r'.*Mercedes-Benz.*': 'Mercedes-Benz',
    r'.*Porsche.*': 'Porsche',
    r'.*Ford.*': 'Ford',
    r'.*Audi.*': 'Audi',
    r'.*Renault.*': 'Renault',
    r'.*BMW.*': 'BMW',
    r'.*Daewoo.*': 'Daewoo',
    r'.*Chevrolet.*': 'Chevrolet',
    r'.*Mazda.*': 'Mazda',
    r'.*Volkswagen.*': 'Volkswagen',
    r'.*Ssangyong.*': 'Ssangyong',
    r'.*Kia.*': 'Kia',
    r'.*Mitsubishi.*': 'Mitsubishi',
    r'.*Subaru.*': 'Subaru',
    r'.*Lincoln.*': 'Lincoln',
    r'.*Fiat.*': 'Fiat',
    r'.*Mini.*': 'Mini',
    r'.*Jeep.*': 'Jeep',
     r'.*Dodge.*': 'Dodge',
     r'.*Skoda.*': 'Skoda',
     r'.*GMC.*': 'GMC',
     r'.*Acura.*': 'Acura',
     r'.*Daihatsu.*': 'Daihatsu',
     r'.*Suzuki.*': 'Suzuki',
    r'.*Buick.*': 'Buick',
    r'.*Jaguar.*': 'Jaguar',
    r'.*Volvo.*': 'Volvo',
    r'.*Rover.*': 'Land Rover',
    r'.*Chrysler.*': 'Chrysler',
    r'.*Isuzu.*': 'Isuzu',
    r'.*Cadillac.*': 'Cadillac',
    r'.*Scion.*': 'Scion',
    r'.*VAZ.*': 'VAZ',
    r'.*UAZ.*': 'UAZ',
    r'.*GAZ.*': 'GAZ',
    r'.*Mercury.*': 'Mercury',
    r'.*Seat.*': 'Seat',
    r'.*Infiniti.*': 'Infiniti',
    r'.*Citroen.*': 'Citroen',
    r'.*Moskvich.*': 'Moskvich',
    r'.*Maserati.*': 'Maserati',
    r'.*Peugeot.*': 'Peugeot',
    r'.*Pontiac.*': 'Pontiac',
     r'.*Greatwall.*': 'Greatwall',
     r'.*Rolls-Royce.*': 'Rolls-Royce',
}

# Replace all patterns in one step
for pattern, replacement in brand_mappings.items():
    df['Brand'] = df['Brand'].str.replace(pattern, replacement, regex=True, case=False)

```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
plt.figure(figsize=(16, 6))
df['Brand'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Number of Cars per Brand')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.histplot(df['price'], kde=True, bins=30)
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(25, 8))
sns.boxplot(x='Brand', y='price', data=df)
plt.title('Car Price Distribution by Car Brand')
plt.xlabel('Brand')
plt.ylabel('Price')
plt.show()
```

```python
most_expensive_brand = df.loc[df.groupby('Brand')['price'].idxmax(), ['Brand', 'price']]
most_expensive_brand
```

```python
most_expensive_brand = df.groupby('Brand')['price'].mean().idxmax()
most_expensive_brand
```

```python
plt.figure(figsize=(16, 6))
df['ManufacturingYear'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Number of Cars per Brand')
plt.xlabel('ManufacturingYear')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(25, 8))
sns.boxplot(x='Brand', y='CylinderCount', data=df)
plt.title('Number of Cylinders by Car Type')
plt.xlabel('Car Type')
plt.ylabel('Cylinders')
plt.show()
```
